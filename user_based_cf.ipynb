```python
import pandas as pd
import numpy as np
```

```python
movies= pd.read_csv('/content/movie.csv')
```

```python
ratings= pd.read_csv('/content/rating.csv')
```

```python
movies
```

```python
ratings
```

```python
movie_rating=movies.merge(ratings,on='movieId')
```

```python
movie_rating
```

```python
movie_rating=movie_rating.drop_duplicates('title')
```

```python
movie_rating
```

```python
movie_rating=movie_rating.drop(columns='timestamp')
```

```python
movie_rating
```

```python
#considering the users that have rated more than 50 movies
x = movie_rating.groupby('userId').count()['rating'] > 50
p_users= x[x].index#Boolean indexing
```

```python
filtered_rating = movie_rating[movie_rating['userId'].isin(p_users)]
```

```python
filtered_rating
```

```python
filtered_rating.drop_duplicates()
```

```python
pt=filtered_rating.pivot_table(index='title',columns='userId',values='rating')
```

```python
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score
```

```python
similarity_score.shape# har movie ka har movie ke saath sim score
```

```python
def recommend(movie_name):
    #fetch index from movie name
    index=np.where(pt.index ==movie_name)[0][0]
    similar_items= sorted(list(enumerate(similarity_score[index])),key=lambda x :x[1],reverse= True)[1:6]
    for i in similar_items:
        print(pt.index[i[0]])
```

```python
recommend('[REC] (2007)')
```

```python
recommend('xXx: State of the Union (2005)')
```
